## 1. My lane as an ML task (type)

*Classification, clustering, ranking, or scoring — which one, and why?*



The selected lane, **Refresh / Content Opportunity Scoring**, is primarily a **Ranking / Scoring** machine learning task. The objective is not simply to predict whether a content page should be refreshed, but to prioritize content pages based on their likelihood of benefiting from a review.

Rather than producing a binary yes/no decision, the expected output is a **priority score** that allows content teams to rank pages from highest to lowest review priority. This enables editors to focus their limited time and resources on the pages that are most likely to require attention.

Although a classification model may be used internally to estimate the probability that a page is declining, the final business objective is a ranked list of content pages. Therefore, the overall problem is best framed as a **Ranking / Scoring** task, where the output directly supports content prioritization and decision-making.

## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*



The ideal business objective is to identify content pages that are most likely to benefit from a refresh or review. However, the starter dataset does not contain a direct label indicating whether refreshing a page will improve its future performance.

Instead, I will use **`is_declining_label`** as the initial **proxy target**. This label indicates whether a content page is currently experiencing declining performance and serves as a reasonable indicator of pages that may require attention.

Using a proxy target allows the model to learn meaningful patterns from the available data while supporting the overall goal of prioritizing content for review. In a production environment, a stronger target could be developed using historical refresh actions and future performance data to directly measure the impact of content updates.

## 3. Success metric

*One metric you can defend. What number means 'good'?*


The primary objective of this project is to produce a ranked list of content pages that helps content teams prioritize their review efforts. Since editors typically review only the highest-priority pages, the evaluation metric should reflect the quality of those recommendations rather than overall prediction accuracy.

The primary success metric for this project is **Precision@K**, which measures how many of the top-ranked pages actually require review or refresh. A high Precision@K indicates that the model is successfully identifying the pages that are most valuable for editorial attention.

In addition to Precision@K, supporting metrics such as **Average Precision** and **Recall** can be used to evaluate the overall quality of the ranking and the model's ability to identify pages that genuinely require attention.

Overall, success is defined by the model's ability to prioritize the right pages for review, enabling content teams to make efficient and informed decisions with limited resources.

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*


A simple rule-based system could prioritize content pages using manually defined thresholds, such as refreshing pages with low CTR or content that has not been updated for a long time. While these rules may work for straightforward situations, they cannot effectively capture the complex relationships between multiple content and search performance signals.

In this project, the decision to prioritize a page depends on the combined influence of several factors, including impressions, clicks, sessions, average search position, content age, days since the last update, competition level, content type, and search intent. The importance of these factors may vary from one page to another, making it difficult to define a single set of rules that performs well in every situation.

A machine learning approach can learn these relationships directly from historical data and assign a priority score based on the interaction of multiple features. This enables the system to produce a more accurate and flexible ranking of content pages while still allowing content editors to make the final review and publishing decisions.

Therefore, machine learning provides a more scalable and reliable solution for content prioritization than relying solely on manually defined rules.
